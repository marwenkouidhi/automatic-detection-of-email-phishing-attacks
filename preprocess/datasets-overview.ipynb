```python
from pymongo import MongoClient
import pandas as pd
# Connect to MongoDB database
client = MongoClient('mongodb://localhost:27017/')
db = client['pidb']

# Retrieve data from MongoDB collection
nazario_trec07_headers_data = list(db['emails_headers'].find({}, {"_id": 0}))
nazario_trec07_headers = pd.DataFrame(nazario_trec07_headers_data)

nazario_trec07_body_data = list(db['emails_bodies'].find({}, {"_id": 0}))
nazario_trec07_body = pd.DataFrame(nazario_trec07_body_data)

phishload_data = list(db['urls'].find({}, {"_id": 0}))
phishload = pd.DataFrame(phishload_data)

liar_data = list(db['fake_news'].find({}, {"_id": 0}))
liar = pd.DataFrame(liar_data)

cyberbullying_data = list(db['cyberbullying'].find({}, {"_id": 0}))
cyberbullying = pd.DataFrame(cyberbullying_data)

# Close MongoDB connection
client.close()
```

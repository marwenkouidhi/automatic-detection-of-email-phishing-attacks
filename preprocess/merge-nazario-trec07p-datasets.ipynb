```python
import pandas as pd 
pd.set_option('display.max_colwidth', None)
```

### Importing the Nazario dataset

```python
nazario_mbox = pd.read_csv('../datasets/nazario-phishing-corpus/csv/nazario-mbox.csv')
nazario_text = pd.read_csv('../datasets/nazario-phishing-corpus/csv/nazario-text.csv')
nazario = pd.concat([nazario_mbox, nazario_text], ignore_index=True)
```

### Importing the trec07 dataset

```python
trec07p = pd.read_csv('../datasets/trec07p/trec07p.csv')
trec07p = trec07p[trec07p.label == 0 ]
```

```python
trec07p
```

```python
with open('features.txt', 'r') as f:
    columns = [line.strip() for line in f.readlines()]

df = pd.concat([nazario, trec07p], ignore_index=True).reindex(columns=columns)
df
```

```python
df.to_csv('../datasets/nazario_trec07p.csv', index=False)
```

```python
headers_columns = df.columns.to_list()
headers_columns.remove('body')

text_columns = ['body', 'label']

nazario_trec07_headers = df[headers_columns]
nazario_trec07_text = df[text_columns]
nazario_trec07_headers.to_csv('../datasets/nazario_trec07_headers.csv', index=False)
nazario_trec07_text.to_csv('../datasets/nazario_trec07_text.csv', index=False)
```
